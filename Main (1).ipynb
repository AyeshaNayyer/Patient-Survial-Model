```python
import pandas as pd

# Specify the path to your CSV fila
train = '/content/drive/MyDrive/train.csv'  # Replace with the actual path to your CSV file
# Read the CSV file into a DataFrame
df = pd.read_csv(train)
test = '/content/drive/MyDrive/test.csv'
df1 = pd.read_csv(test)



```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['ethnicity','gender','icu_admit_source','icu_stay_type','icu_type','apache_3j_bodysystem','apache_2_bodysystem'])
```

```python
# impute missing values via the column mean

from sklearn.impute import KNNImputer
import numpy as np
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
numerical_cols1 = df1.select_dtypes(include=[np.number]).columns.tolist()

# Compute the means for numerical columns and impute missing values
for col in numerical_cols:
    imputer = KNNImputer(n_neighbors=5)
    df[col] = imputer.fit_transform(df[[col]])
for col in numerical_cols1:
    df1[col] = imputer.fit_transform(df1[[col]])
```

```python
# Putting feature variable to X
X_train = df.drop('hospital_death',axis=1)
# Putting response variable to y
y_train = df['hospital_death']
X_test = df1
```

```python
# To store results of models
result_dict_train = {}
result_dict_test = {}
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

# Reduce the parameter grid and n_iter
param_dist = {
    'learning_rate': [0.01, 0.05, 0.1],
    'n_estimators': np.arange(100, 500, 100),
    'max_depth': [3, 4],  # Reduced range
    'min_samples_split': [2, 5],  # Reduced range
}

# Create the GradientBoostingRegressor
gb = GradientBoostingRegressor()

# Create RandomizedSearchCV with fewer iterations and parallel processing
gb_random = RandomizedSearchCV(
    gb, param_distributions=param_dist, n_iter=40, cv=4, random_state=69, n_jobs=-1)

# Use a smaller sample of your training data for faster experimentation
sample_size = 1000  # Adjust the sample size as needed
X_train_sample = X_train[:sample_size]
y_train_sample = y_train[:sample_size]

# Fit RandomizedSearchCV to the sample data
gb_random.fit(X_train_sample, y_train_sample)

# Print the best parameters and scores
print("Best Parameters:", gb_random.best_params_)
print("Train Score (CV):", gb_random.best_score_)

# Evaluate the model on the full test set
test_score = gb_random.score(X_test, y_test)
print("Test Score:", test_score)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV
import numpy as np

# Choose all predictors except target & IDcols

param_test1 = {'n_estimators': range(20, 81, 10)}

gsearch1 = GridSearchCV(
    estimator=GradientBoostingClassifier(
        learning_rate=0.1,
        min_samples_split=500,
        min_samples_leaf=50,
        max_depth=8,
        max_features='sqrt',
        subsample=0.8,
        random_state=10
    ),
    param_grid=param_test1,  # Note that this is param_grid, not param_test1
    scoring='roc_auc',
    n_jobs=4,
    cv=5
)

gsearch1.fit(X_train, y_train)  # I assume you meant to use X_train and y_train here
```

```python
test_probabilities = gsearch1.predict_proba(X_test)  # Get probabilities for the positive class

# Load the test data that contains the "RecordID" column
test_file_path = '/content/drive/MyDrive/test.csv'
test_data = pd.read_csv(test_file_path)

# Create a DataFrame with the "RecordID" and predicted probability
predictions_df = pd.DataFrame({
    "RecordID": test_data["RecordID"],
    "hospital_death_probability": test_probabilities[:, 1]
})

# Save the predictions to a CSV file
predictions_df.to_csv('test_predictions27.csv', index=False)






```
